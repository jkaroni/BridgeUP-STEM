First, we  import the necessary libraries:

```python
import matplotlib.pyplot as plt
import numpy as np
import csv 
```

```python
csvFile = open("CloudFraction1_2017.csv", 'r') # r means 'read only' 
csvReader = csv.reader(csvFile, delimiter=',') #library function as it is a function from in this case csv 
#says that a ',' separates the components in the list
data_lists = list(csvReader) #turns into a datalist
```

```python
width = len(data_lists[0])
height = len(data_lists)
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0.1:
            grid[r, c] = [99, 202, 255]
        elif float(value) <= 0.2:
            grid[r, c] = [119, 209, 255]
        elif float(value) <= 0.3:
            grid[r, c] = [127, 212, 255]
        elif float(value) <= 0.4:
            grid[r, c] = [142, 217, 255]
        elif float(value) <= 0.5:
            grid[r, c] = [160, 224, 255]
        elif float(value) <= 0.6:
            grid[r, c] = [181, 231, 255]
        elif float(value) <= 0.7:
            grid[r, c] = [201, 238, 255]
        elif float(value) <= 0.8:
            grid[r, c] = [226, 246, 255]
        elif float(value) <= 0.9:
            grid[r, c] = [249, 253, 255]
        elif float(value) <= 1.0:
            grid[r, c] = [255, 255, 255]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
plt.title("Global sea temperature in Celsius")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Sea_temperature_2017.png",dpi = 1000)
plt.show()
```
